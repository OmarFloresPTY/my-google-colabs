```python
fig, axs = plt.subplots(2,2) #Crea una figura y un subconjunto de subfiguras con una matriz de 2x2
x = np.linspace(1,10,100)
y = np.linspace(1,10,100)
axs[0,0].plot(x, y)
axs[0,0].set_title('Gráfica Linear')
axs[0,1].plot(x**2,y)
axs[0,1].set_title('Gráfica Cuadratica')
axs[1,1].plot(x**3,y)
axs[1,1].set_title('Gráfica Cúbica')
axs[1,0].plot(x,np.sin(x))
axs[1,0].set_title('Coseno')
plt.tight_layout() #Esto solo es para que no se superpongan las gráficas.
```

```python
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])

axs['right_top'].plot(x, np.cos(x), 'r', label='y = cos(x)')
axs['right_top'].set_title('Gráfica Coseno')

axs['left'].plot(x, np.sin(x), 'g', label='y = sen(x)')
axs['left'].set_title('Gráfica Seno')

axs['right_bottom'].plot(x, x**2, 'm', label='y = tan(x)')
axs['right_bottom'].set_title('Gráfica Cuadrática')

axs['left'].legend()
axs['right_top'].legend()
axs['right_bottom'].legend()
plt.tight_layout()
plt.show()
```

```python
# Gráfica en el primer subplot (izquierda)
axs['left'].plot(x, y, label='y = sin(x)')
axs['left'].set_title('Gráfica de y = sin(x)')

# Gráfica en el segundo subplot (arriba a la derecha)
axs['right_top'].plot(x, y, 'r', label='y = cos(x)')
axs['right_top'].set_title('Gráfica de y = cos(x)')

# Gráfica en el tercer subplot (abajo a la izquierda)
axs['left'].plot(x, y, 'g', label='y = exp(-x/5) * sin(2*x)')
axs['left'].set_title('Gráfica de y = exp(-x/5) * sin(2*x)')

# Gráfica en el cuarto subplot (abajo a la derecha)
axs['right_bottom'].plot(x, y, 'm', label='y = exp(-x/5)')
axs['right_bottom'].set_title('Gráfica de y = exp(-x/5)')

# Añadir leyendas
axs['left'].legend()
axs['right_top'].legend()
axs['right_bottom'].legend()
```

##Tipos de entrada o input a gráficar.

```python
b = np.matrix([[1,2],[3,4]])
b_asarray = np.asanyarray(b)
```

```python
data = {
    'a':np.arange(50),
    'c':np.random.randint(0,50,50),
    'd':np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d'])*100
```

```python
fig, ax = plt.subplots(figsize=(5,2.7),layout='constrained')
ax.scatter('a','b', c ='c', s='d',data=data) #Crea un gráfico de dispersión
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```
