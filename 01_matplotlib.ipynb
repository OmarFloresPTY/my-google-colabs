```python
fig, axs = plt.subplots(2,2) #Crea una figura y un subconjunto de subfiguras con una matriz de 2x2
x = np.arange(1,10)
y = np.arange(1,10)
axs[0,0].plot(x, y)
axs[0,0].set_title('Gr√°fica Linear')
axs[0,1].plot(x**2,y)
axs[0,1].set_title('Gr√°fica Cuadratica')
axs[1,1].plot(x**3,y)
axs[1,1].set_title('Gr√°fica C√∫bica')
axs[1,0].plot(x,np.sin(x))
axs[1,0].set_title('Coseno')
plt.tight_layout() #Esto solo es para que no se superpongan las gr√°ficas.
```

```python
fig, axs = plt.subplot_mosaic([['left', 'right_top'],
                               ['left', 'right_bottom']])
```

##Tipos de entrada o input a gr√°ficar.

```python
b = np.matrix([[1,2],[3,4]])
b_asarray = np.asanyarray(b)
```

```python
data = {
    'a':np.arange(50),
    'c':np.random.randint(0,50,50),
    'd':np.random.randn(50)
}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d'])*100
```

```python
fig, ax = plt.subplots(figsize=(5,2.7),layout='constrained')
ax.scatter('a','b', c ='c', s='d',data=data) #Crea un gr√°fico de dispersi√≥n
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```
