##Dibuja el primer plot

```python
x = np.linspace(0,2*np.pi,200)
y = np.sin(x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots() #Crea figura simples de dos ejes (x,y)
ax.plot([1,2,3,4],[1,4,2,3])
#plt.show() #En un cuaderno de jupyter no hace falta llamar a este método.
```
